```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("superstore.db")
```

```python
df = pd.read_csv(r"C:\Users\shokd\Desktop\Superstore\sales-revenue-intelligence\data\train.csv")
```

```python
df.to_sql(
    "sales",
    conn,
    if_exists="replace",
    index=False
)
```

```python
pd.read_sql_query(
    "SELECT name FROM sqlite_master WHERE type='table';",
    conn
)
```

```python
pd.read_sql_query(
    "PRAGMA table_info(sales);",
    conn
)
```

```python
query = """
SELECT
    STRFTIME('%Y', "Order Date") AS order_year,

    ROUND(SUM("Sales"), 2) AS total_revenue,

    ROUND(
        (
            SUM("Sales") -
            LAG(SUM("Sales")) OVER (
                ORDER BY STRFTIME('%Y', "Order Date")
            )
        ) * 100.0 /

        LAG(SUM("Sales")) OVER (
            ORDER BY STRFTIME('%Y', "Order Date")
        ),
        2
    ) AS yoy_growth_pct

FROM sales

GROUP BY STRFTIME('%Y', "Order Date")

ORDER BY order_year;
"""

result = pd.read_sql_query(query, conn)

result
```

```python
pd.read_sql_query(
    '''
    SELECT "Order Date"
    FROM sales
    LIMIT 5;
    ''',
    conn
)
```

```python
query = """
SELECT
    SUBSTR("Order Date", -4) AS order_year,

    ROUND(SUM("Sales"), 2) AS total_revenue,

    ROUND(
        (
            SUM("Sales") -
            LAG(SUM("Sales")) OVER (
                ORDER BY SUBSTR("Order Date", -4)
            )
        ) * 100.0 /

        LAG(SUM("Sales")) OVER (
            ORDER BY SUBSTR("Order Date", -4)
        ),
        2
    ) AS yoy_growth_pct

FROM sales

GROUP BY SUBSTR("Order Date", -4)

ORDER BY order_year;
"""

result = pd.read_sql_query(query, conn)

result
```

```python
query = """
SELECT

    SUBSTR("Order Date", 4, 2) AS order_month,

    ROUND(SUM("Sales"), 2) AS monthly_revenue

FROM sales

GROUP BY order_month

ORDER BY order_month;
"""

monthly_sales = pd.read_sql_query(query, conn)

monthly_sales
```

```python
query = """
SELECT
    "Region",
    ROUND(SUM("Sales"), 2) AS total_revenue
FROM sales
GROUP BY "Region"
ORDER BY total_revenue DESC;
"""

result = pd.read_sql_query(query, conn)

result
```
